```python
# Neon Green-Pink Color Codes
RESET = "\033[0m"
NEON_GREEN = "\033[92m"
NEON_PINK = "\033[95m"
NEON_YELLOW = "\033[93m"
NEON_RED = "\033[91m"

contacts = []

def add_contact():
    print(f"{NEON_GREEN}-- Add New Contact --{RESET}")
    name = input("Name: ")
    phone = input("Phone: ")
    email = input("Email: ")
    address = input("Address: ")
    for c in contacts:
        if c['name'].lower() == name.lower():
            print(f"{NEON_RED}‚ùå Contact already exists!{RESET}")
            return
    contacts.append({
        "name": name,
        "phone": phone,
        "email": email,
        "address": address
    })
    print(f"{NEON_PINK}‚úÖ Contact added successfully!{RESET}")

def view_contacts():
    print(f"{NEON_GREEN}-- All Contacts --{RESET}")
    if not contacts:
        print(f"{NEON_YELLOW}No contacts yet.{RESET}")
        return
    for i, c in enumerate(contacts, 1):
        print(f"{NEON_PINK}{i}. {c['name']}{RESET}")
        print(f"   üìû {c['phone']}")
        print(f"   üìß {c['email']}")
        print(f"   üè† {c['address']}\n")

def search_contact():
    query = input("Search name or phone: ").lower()
    found = False
    for c in contacts:
        if query in c['name'].lower() or query in c['phone']:
            print(f"{NEON_PINK}Found:{RESET}")
            print(f"Name: {c['name']}")
            print(f"Phone: {c['phone']}")
            print(f"Email: {c['email']}")
            print(f"Address: {c['address']}")
            found = True
    if not found:
        print(f"{NEON_RED}‚ùå No match found.{RESET}")

def delete_contact():
    name = input("Name to delete: ")
    global contacts
    for c in contacts:
        if c['name'].lower() == name.lower():
            contacts = [x for x in contacts if x['name'].lower() != name.lower()]
            print(f"{NEON_PINK}üóëÔ∏è Contact '{name}' deleted.{RESET}")
            return
    print(f"{NEON_RED}‚ùå Contact not found.{RESET}")

def menu():
    while True:
        print(f"""
{NEON_GREEN}============================
    NEON CONTACT BOOK
============================{RESET}
1. Add Contact
2. View All Contacts
3. Search Contact
4. Delete Contact
0. Exit
""")
        choice = input("Select option: ")
        if choice == "1":
            add_contact()
        elif choice == "2":
            view_contacts()
        elif choice == "3":
            search_contact()
        elif choice == "4":
            delete_contact()
        elif choice == "0":
            print(f"{NEON_YELLOW}Goodbye!{RESET}")
            break
        else:
            print(f"{NEON_RED}Invalid option!{RESET}")

# Run the menu
menu()
```
